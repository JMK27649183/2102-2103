```python
class Queue:
    def __init__(self):
        self.items = []

    def isEmpty(self):
        return self.items == []

    def enqueue(self, item):
        self.items.insert(0,item)

    def dequeue(self):
        if len(self.items) > 0 :
            return self.items.pop()
        else : 
            return ''

    def size(self):
        return len(self.items)
    
    def peek(self) :
        if len(self.items) > 0 :
            return self.items[-1]

        else : 
            return '*'

    def peek1(self) :
        if len(self.items) > 1 :
            return self.items[-2]

        else : 
            return '*'
```

```python
def chemf(k):

    a = Queue()
    b = Queue()
    dic = {}

    for i in k:
        a.enqueue(i)

    while not a.isEmpty():
        
        x = a.dequeue()
        
        if x == '(':
        
            t = a.dequeue()
            
            while True:
                
                b.enqueue(t)
                t = a.dequeue()
                
                if t == ')':
                    break
            
            m = ''
            
            while a.peek() in '1234567890':
                m = str(m) + str(a.dequeue())
                m = int(m)
                
            while not b.isEmpty():
                
                w = b.dequeue()
                
                if b.peek() in 'ABCDEFGHIJKLMNOPQRSTUVWXYZ':
                
                    if w in dic:
                        dic[w] = dic[w] + m
                        
                    else:
                        dic[w] = m

                elif b.peek() in '1234567890':
                    
                    q = ''
                    
                    while b.peek() in '1234567890':
                        q += b.dequeue()
                    
                    q = int(q)
                    
                    if w in dic:
                        dic[w] += q * m
                        
                    else:
                        dic[w] = q * m

                elif b.peek() == '*':
                    
                    if w in dic:
                        dic[w] = dic[w] + m
                        
                    else:
                        dic[w] = m

                else:
                    
                    l = b.dequeue()
                    u = w + l
                    
                    if b.peek() in 'ABCDEFGHIJKLMNOPQRSTUVWXYZ':
                        
                        if u in dic:
                            dic[u] = dic[u] + m
                            
                        else:
                            dic[u] = m

                    elif b.peek() in '1234567890':
                        
                        y = ''
                        
                        while b.peek() in '1234567890':
                            y += b.dequeue()
                            
                        y = int(y)
                        
                        if x in dic:
                            dic[x] += y * m
                            
                        else:
                            dic[x] = y * m

                    else:
                        
                        if u in dic:
                            dic[u] = dic[u] + m
                            
                        else:
                            dic[u] = m

        else:
            
            if a.peek() in 'ABCDEFGHIJKLMNOPQRSTUVWXYZ':
                
                if x in dic:
                    dic[x] = dic[x] + 1
                
                else:
                    dic[x] = 1

            elif a.peek() in '1234567890':
                
                y = ''
                
                while a.peek() in '1234567890':
                    y += a.dequeue()

                y = int(y)
                
                if x in dic:
                    dic[x] += y
                
                else:
                    dic[x] = y

            elif a.peek() == '*':
               
                if x in dic:
                    dic[x] = dic[x] + 1
                
                else:
                    dic[x] = 1

            elif a.peek() in 'abcdefghijklmnopqrstuvwxyz':
                
                e = a.dequeue()
                z = x + e

                if a.peek() in 'ABCDEFGHIJKLMNOPQRSTUVWXYZ':
                
                    if z in dic:
                        dic[z] = int(dic[z]) + 1

                    else:
                        dic[z] = 1

                elif a.peek() in '1234567890':
                    
                    y = ''
                    
                    while a.peek() in '1234567890':
                        y += a.dequeue()
                    
                    y = int(y)
                    
                    if z in dic :
                        dic[z] = int(dic[z]) + y
                    
                    else:
                        dic[z] = y

                else:
                    
                    if z in dic:
                        dic[z] = dic[z] + 1
                    
                    else:
                        dic[z] = 1
    
    return dic
```

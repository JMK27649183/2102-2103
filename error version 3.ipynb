```python
class Queue:
    def __init__(self):
        self.items = []

    def isEmpty(self):
        return self.items == []

    def enqueue(self, item):
        self.items.insert(0,item)

    def dequeue(self):
        if len(self.items) > 0 :
            return self.items.pop()
        else : 
            return ''

    def size(self):
        return len(self.items)
    
    def peek(self) :
        if len(self.items) > 0 :
            return self.items[0]

        else : 
            return '*'

    def peek1(self) :
        if len(self.items) > 1 :
            return self.items[1]

        else : 
            return '*'
```
